```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
X_train = train[['family_structure','location','gross_income']].as_matrix()
y_train = train[['rate']].as_matrix()
X_test = test[['family_structure','location','gross_income']].as_matrix()
y_test = test[['rate']].as_matrix()
```

```python
svr = SVR(kernel='linear',epsilon=0.1)
svr.fit(X_train,y_train)
```

```python
plt.plot(np.arange(0,24.7,0.1),train['gross_income'].as_matrix(),'*')
plt.title("Data distribution")
plt.show()
```

```python
print (svr.score(X_test,y_test))
y_test.sort()
```

```python
x_axis = np.arange(0,8.3,0.1)
y_test = y_test.reshape(1,83)
y_test.sort()
y_predict = svr.predict(X_test)
y_predict.sort()
plt.plot(x_axis,y_test[0],'*',x_axis,y_predict,'.')
plt.title("Result")
plt.show()
```
